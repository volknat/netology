```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
import pandas as pd
```

```python
features=pd.DataFrame(iris['data'])
```

```python
target=pd.DataFrame(iris['target'])
```

```python
target['target']=target[0]
target.drop(0, axis=1, inplace=True)
```

```python
f_t=pd.concat([features, target], axis=1)
```

```python
f_t=f_t[f_t['target']!=0]
```

```python
f_t['target']=[x if x!=2 else x-2 for x in f_t['target']]
```

```python
Y=f_t['target']
X=f_t.drop('target', axis=1)
print(X)
```

```python
LEARNING_RATE = 0.001

costs = []
params = []
preds = []
params = np.random.normal(size=(5,))
for i in range(2):
    prob =1/( 1+np.exp(-(np.dot(params[1:], X.T)+params[0])))
    preds.append(prob)
    cost = -sum(Y*(np.dot(params[1:], X.T)+params[0])-np.log(1+np.exp(np.dot(params[1:], X.T)+params[0])))
    costs.append(cost)
    params[0] -= -LEARNING_RATE *np.sum(Y-prob)
    params[1] -= -LEARNING_RATE *np.sum((Y-prob)*X[0])
    params[2] -= -LEARNING_RATE *np.sum((Y-prob)*X[1])
    params[3] -= -LEARNING_RATE *np.sum((Y-prob)*X[2])
    params[4] -= -LEARNING_RATE *np.sum((Y-prob)*X[3])
```

```python
LEARNING_RATE = 0.001
gamma=0.1

costs = []
params= []
preds= []
params = np.random.normal(size=(5,))
grad_0=0
grad_1=0
grad_2=0
grad_3=0
grad_4=0

prob =1/( 1+np.exp(-(np.dot(params[1:], X.T)+params[0])))
preds.append(prob)
cost = -sum(Y*(np.dot(params[1:], X.T)+params[0])-np.log(1+np.exp(np.dot(params[1:], X.T)+params[0])))
costs.append(cost)
grad_0=-LEARNING_RATE *np.sum(Y-prob)
grad_1=-LEARNING_RATE *np.sum((Y-prob)*X[0])
grad_2=-LEARNING_RATE *np.sum((Y-prob)*X[1])
grad_3=-LEARNING_RATE *np.sum((Y-prob)*X[2])
grad_4=-LEARNING_RATE *np.sum((Y-prob)*X[3])
params[0] -= grad_0
params[1] -= grad_1
params[2] -= grad_2
params[3] -= grad_3
params[4] -= grad_4
prob =1/( 1+np.exp(-(np.dot(params[1:], X.T)+params[0])))
preds.append(prob)
cost = -sum(Y*(np.dot(params[1:], X.T)+params[0])-np.log(1+np.exp(np.dot(params[1:], X.T)+params[0])))
costs.append(cost)
grad_0=-LEARNING_RATE *np.sum(Y-prob)+grad_0*gamma
grad_1=-LEARNING_RATE *np.sum((Y-prob)*X[0])+grad_1*gamma
grad_2=-LEARNING_RATE *np.sum((Y-prob)*X[1])+grad_2*gamma
grad_3=-LEARNING_RATE *np.sum((Y-prob)*X[2])+grad_3*gamma
grad_4=-LEARNING_RATE *np.sum((Y-prob)*X[3])+grad_4*gamma
params[0] -= grad_0
params[1] -= grad_1
params[2] -= grad_2
params[3] -= grad_3
params[4] -= grad_4
```

```python
while costs[len(costs)-2]-costs[len(costs)-1]>0.001:
    prob =1/( 1+np.exp(-(np.dot(params[1:], X.T)+params[0])))
    preds.append(prob)
    cost = -sum(Y*(np.dot(params[1:], X.T)+params[0])-np.log(1+np.exp(np.dot(params[1:], X.T)+params[0])))
    costs.append(cost)
    grad_0=-LEARNING_RATE *np.sum(Y-prob)+grad_0*gamma
    grad_1=-LEARNING_RATE *np.sum((Y-prob)*X[0])+grad_1*gamma
    grad_2=-LEARNING_RATE *np.sum((Y-prob)*X[1])+grad_2*gamma
    grad_3=-LEARNING_RATE *np.sum((Y-prob)*X[2])+grad_3*gamma
    grad_4=-LEARNING_RATE *np.sum((Y-prob)*X[3])+grad_4*gamma
    params[0] -= grad_0
    params[1] -= grad_1
    params[2] -= grad_2
    params[3] -= grad_3
    params[4] -= grad_4
```
